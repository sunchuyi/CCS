```python
#Read in data

import pandas as pd
import collections as co
import math
import numpy as np
import scipy.sparse as spr
import nltk

bios = pd.read_csv("bios.csv", usecols = ["Bio","Label","Year","Name"])
```

```python
#Clean up
n = bios.shape[0]

bios.Bio = bios.Bio.str.lower()

bios.Name = bios.Name.str.lower()

for i in range(n):
    bios.Bio.loc[i] = nltk.word_tokenize(bios.Bio.loc[i])
    bios.Name.loc[i] = nltk.word_tokenize(bios.Name.loc[i])

for i in range(n):
    if (bios.Label.loc[i] != 'F') or (bios.Label.loc[i] != 'M'):
        if ('F' in bios.Label.loc[i]):
            bios.loc[i,"Label"] = 'f'
        elif ('M' in bios.Label.loc[i]):
            bios.loc[i,"Label"] = 'm'
```

```python
#document frequency, with tags

for i in range(n):
    bios.Bio.loc[i] = nltk.pos_tag(bios.Bio.loc[i])

dict_base = co.Counter()
    
dict_nn = co.Counter()
dict_jj = co.Counter()
dict_vb = co.Counter()
dict_rb = co.Counter()

dict_nn_m = co.Counter()
dict_jj_m = co.Counter()
dict_vb_m = co.Counter()
dict_rb_m = co.Counter()

dict_nn_f = co.Counter()
dict_jj_f = co.Counter()
dict_vb_f = co.Counter()
dict_rb_f = co.Counter()

for j in range(len(bios.Bio.loc[i])):
    if (bios.Bio.loc[i][j][0] not in bios.Name.loc[i]) and (bios.Bio.loc[i][j][0].isalpha()):
        dict_base[bios.Bio.loc[i][j][0]] += 1

for i in range(n):
    docflag = dict_base.copy()
    if bios.Label.loc[i] == 'm':
        for j in range(len(bios.Bio.loc[i])):
            if (bios.Bio.loc[i][j][0] not in bios.Name.loc[i]) and (bios.Bio.loc[i][j][0].isalpha()):
                if (docflag[bios.Bio.loc[i][j][0]] > -1):
                    docflag[bios.Bio.loc[i][j][0]] = -1
                    if 'NN' in bios.Bio.loc[i][j][1]:
                        dict_nn[bios.Bio.loc[i][j][0]] += 1
                        dict_nn_m[bios.Bio.loc[i][j][0]] += 1
                    elif 'JJ' in bios.Bio.loc[i][j][1]:
                        dict_jj[bios.Bio.loc[i][j][0]] += 1
                        dict_jj_m[bios.Bio.loc[i][j][0]] += 1
                    elif 'VB' in bios.Bio.loc[i][j][1]:
                        dict_vb[bios.Bio.loc[i][j][0]] += 1
                        dict_vb_m[bios.Bio.loc[i][j][0]] += 1
                    elif 'RB' in bios.Bio.loc[i][j][1]:
                        dict_rb[bios.Bio.loc[i][j][0]] += 1
                        dict_rb_m[bios.Bio.loc[i][j][0]] += 1
                
    else:
        for j in range(len(bios.Bio.loc[i])):
            if (bios.Bio.loc[i][j][0] not in bios.Name.loc[i]) and (bios.Bio.loc[i][j][0].isalpha()):
                if (docflag[bios.Bio.loc[i][j][0]] > -1):
                    docflag[bios.Bio.loc[i][j][0]] = -1
                    if 'NN' in bios.Bio.loc[i][j][1]:
                        dict_nn[bios.Bio.loc[i][j][0]] += 1
                        dict_nn_f[bios.Bio.loc[i][j][0]] += 1
                    elif 'JJ' in bios.Bio.loc[i][j][1]:
                        dict_jj[bios.Bio.loc[i][j][0]] += 1
                        dict_jj_f[bios.Bio.loc[i][j][0]] += 1
                    elif 'VB' in bios.Bio.loc[i][j][1]:
                        dict_vb[bios.Bio.loc[i][j][0]] += 1
                        dict_vb_f[bios.Bio.loc[i][j][0]] += 1
                    elif 'RB' in bios.Bio.loc[i][j][1]:
                        dict_rb[bios.Bio.loc[i][j][0]] += 1
                        dict_rb_f[bios.Bio.loc[i][j][0]] += 1
```

```python
print("Top 10 nouns:")
print(pd.DataFrame(dict_nn.most_common()[:10]))
print()
print("Top 10 nouns for males:")
print(pd.DataFrame(dict_nn_m.most_common()[:10]))
print()
print("Top 10 nouns for females:")
print(pd.DataFrame(dict_nn_f.most_common()[:10]))
```
